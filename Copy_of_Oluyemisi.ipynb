```python
from torch.utils.data import Dataset

class QADataset(Dataset):
    def __init__(self, questions, answers, tokenizer, max_length=512):
        self.questions = questions
        self.answers = answers
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.questions)

    def __getitem__(self, idx):
        question = self.questions.iloc[idx]
        answer = self.answers.iloc[idx]

        # Handle potential NaN values in questions or answers
        if not question:
            question = ""  # Replace with an empty string or another appropriate value
        if not answer:
            answer = ""  # Replace with an empty string or another appropriate value

        encoding = self.tokenizer.encode_plus(
            question,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        decoding = self.tokenizer.encode_plus(
            answer,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': decoding['input_ids'].flatten()
        }
```

```python
# Create train and validation datasets
train_dataset = QADataset(train_df['question'], train_df['answer'], tokenizer)
val_dataset = QADataset(val_df['question'], val_df['answer'], tokenizer)
test_dataset = QADataset(test_df['question'], test_df['answer'], tokenizer)
```

```python
# Create DataLoader instances
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)
```

```python

# Instantiate the T5 model
model = T5ForConditionalGeneration.from_pretrained('t5-base')
# Check if GPU is available, if not, use CPU
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
    print("WARNING: Using CPU as CUDA is not available. Training will be slower.")

# Move the model to the selected device
model.to(device)
```

```python
from transformers import AdamW, get_linear_schedule_with_warmup
from sklearn.metrics import f1_score, accuracy_score, precision_score,recall_score,confusion_matrix

# Define the optimizer and training parameters
optimizer = AdamW(model.parameters(), lr=1e-4)
num_epochs = 10

# Learning rate scheduler
total_steps = len(train_loader) * num_epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Training loop
for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        loss.backward()
        optimizer.step()
        scheduler.step()  # Update learning rate

        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{num_epochs}] Train Loss: {avg_train_loss:.4f}")

# Validation loop
    model.eval()
    with torch.no_grad():
        total_val_loss = 0
        all_predicted_labels = []
        all_true_labels = []

        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_val_loss += loss.item()

            # For F1-score calculation:
            predicted_labels = torch.argmax(outputs.logits, dim=-1).cpu().numpy().flatten()
            true_labels = batch['labels'].cpu().numpy().flatten()
            all_predicted_labels.extend(predicted_labels)
            all_true_labels.extend(true_labels)

        avg_val_loss = total_val_loss / len(val_loader)
        print(f"Epoch [{epoch+1}/{num_epochs}] Validation Loss: {avg_val_loss:.4f}")

# Calculate and print F1-score for validation set
f1 = f1_score(all_true_labels, all_predicted_labels, average='weighted')
print(f"Epoch [{epoch+1}/{num_epochs}] Validation F1-score: {f1:.4f}")

# Save the fine-tuned model
model.save_pretrained('fine_tuned_t5_model')
```

```python

import random
from sklearn.metrics.pairwise import cosine_similarity

# Load the fine-tuned model and tokenizer
model_path = 'fine_tuned_t5_model'  # Path to your saved model
tokenizer = T5Tokenizer.from_pretrained('t5-base')
model = T5ForConditionalGeneration.from_pretrained(model_path)

# Device (CPU or GPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Question bank
question_bank = [
    "What are the entry requirements for the Master in Data Science program?",
    "How long does the International MBA (IMBA) program take to complete?",
    "Is the Master in Entrepreneurship and Innovation program offered online?",
    "What types of scholarships are available for international students?",
    "Are there any financial aid options for the DBA program?",
    "Who are some of the faculty members in the Agribusiness Management department?",
    "What research areas does the E-Research Center focus on?",
    "What documents are required for the application process?",
    "When is the deadline for enrolling in the next intake of the MBA program?",
    "Where is the Rome Business School located?",
    "What career services does the school offer?"
]

# Function to generate embeddings using the T5 model
def generate_embeddings(text_list):
    embeddings = []
    for text in text_list:
        input_ids = tokenizer.encode(text, return_tensors="pt").to(device)
        with torch.no_grad():
            # Get hidden states using encoder output instead of model generation
            outputs = model.encoder(input_ids=input_ids, return_dict=True)
            # Reshape embedding to 2D by squeezing the batch dimension
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze(0).cpu().numpy()
            embeddings.append(embedding)
    return embeddings

# Function to retrieve context using semantic similarity
def get_context(question, data_list):
    # 1. Generate embeddings for the data list and the question
    data_embeddings = generate_embeddings(data_list)
    question_embedding = generate_embeddings([question])[0]  # Get the first element

    # 2. Calculate cosine similarity
    similarities = cosine_similarity(question_embedding.reshape(1, -1), data_embeddings).flatten()
    # Reshape question_embedding for cosine_similarity

    # 3. Get the index of the most similar data point
    most_similar_index = similarities.argmax()

    # 4. Return the context
    context = data_list[most_similar_index]
    return context

# Function to get prediction
def get_prediction(question, data_list):
    start_time = time.time()  #
    context = get_context(question, data_list)
    input_text = f"question: {question} context: {context}"
    input_ids = tokenizer.encode(input_text, return_tensors="pt").to(device)
    outputs = model.generate(input_ids,max_length=512, num_beams=6, early_stopping=False)  # Adjusted parameters)
    prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)
    end_time = time.time()
    response_time = end_time - start_time  # Calculate the response time
    print(f"Response Time: {response_time:.4f} seconds")  # Print the response time
    return prediction

# Assuming you have your scraped data in lists like admissions_process, selection_interview, etc.
# Combine all the scraped data into one list
all_data = admissions_process + selection_interview + scholarships + request + faculty + graduation_ceremony + masters + agric_business + marketing_sales + data_science + entrepreneurship_innovation + logistic_supply + pm + dba + mba + imba + executive_education + courses + tailored_courses + enrollment + research_center + news_events + faq + contact

# Example usage with random question and context retrieval
random_question = random.choice(question_bank)
prediction = get_prediction(random_question, all_data) # Uncomment and replace 'all_data'

print(f"Question: {random_question}")
print(f"Prediction: {prediction}") # Uncomment to print the prediction
```

```python
# Evaluation on the test set (with F1-score)
model.eval()
with torch.no_grad():
    all_predicted_labels = []  # Reset for test set
    all_true_labels = []  # Reset for test set

    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)

        # For F1-score calculation (test set):
        predicted_labels = torch.argmax(outputs.logits, dim=-1).cpu().numpy().flatten()
        true_labels = batch['labels'].cpu().numpy().flatten()
        all_predicted_labels.extend(predicted_labels)
        all_true_labels.extend(true_labels)

    # Calculate and print F1-score for test set
    f1 = f1_score(all_true_labels, all_predicted_labels, average='weighted', zero_division=0)
    accuracy = accuracy_score(all_true_labels, all_predicted_labels)
    precision = precision_score(all_true_labels, all_predicted_labels, average='weighted', zero_division=0)
    recall = recall_score(all_true_labels, all_predicted_labels, average='weighted', zero_division=0)  # Sensitivity
    conf_matrix = confusion_matrix(all_true_labels, all_predicted_labels)

    print(f"Test F1-score: {f1:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")
    print(f"Test Precision: {precision:.4f}")
    print(f"Test Recall (Sensitivity): {recall:.4f}")
    print("Confusion Matrix:")
    print(conf_matrix)
```
